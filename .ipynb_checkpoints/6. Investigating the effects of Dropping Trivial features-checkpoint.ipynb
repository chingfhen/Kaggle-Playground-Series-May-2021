# Randomized search on Catboost with only pair-wise features
- investigate the feature importances of the pairwise features

```python
import random
from catboost import CatBoostClassifier, Pool
import seaborn as sb
from sklearn.model_selection import train_test_split
from sklearn.metrics import log_loss
from collections import defaultdict
```

```python
X = df.iloc[:,:-1]
y = df.target
trainSet,testSet = train_test_split(df, train_size = 0.7, stratify = df.target)
train_X, train_y = trainSet.iloc[:,:-1],trainSet.target
test_X, test_y = testSet.iloc[:,:-1],testSet.target
```

```python
testScores = defaultdict(list)
```

```python
for i in range(5):
    
    # this produces random hyperparameter grids
    random_grid = {'iterations' : 300,
                   'grow_policy' : 'SymmetricTree',
                   'task_type' : 'GPU',
                   'early_stopping_rounds' : 2,
                   'learning_rate': random.sample(list(np.random.uniform(0.1,0.2,20)),1)[0],        
                   'depth': random.sample(list(range(4,11)),1)[0],
                   'l2_leaf_reg': random.sample(list(range(5,15)),1)[0]}
    model = CatBoostClassifier(iterations = random_grid['iterations'],
                               grow_policy = random_grid['grow_policy'],
                               task_type = random_grid['task_type'],
                               early_stopping_rounds = random_grid['early_stopping_rounds'],
                               learning_rate = random_grid['learning_rate'],
                               l2_leaf_reg = random_grid['l2_leaf_reg'])
    model.fit(X,y)
    # get feature importances for that particular grid
    feature_importance_ascending = pd.DataFrame(model.feature_importances_,index = model.feature_names_,columns = ['importance_score']).sort_values(by = 'importance_score').index

    # iteratively ignore more features starting from least important features 
    for num_ignored in range(0,len(X.columns)-2,2):
        ignored_features = list(feature_importance_ascending[:num_ignored])
        model = CatBoostClassifier(iterations = random_grid['iterations'],
                                   grow_policy = random_grid['grow_policy'],
                                   task_type = random_grid['task_type'],
                                   early_stopping_rounds = random_grid['early_stopping_rounds'],
                                   learning_rate = random_grid['learning_rate'],
                                   l2_leaf_reg = random_grid['l2_leaf_reg'],
                                   ignored_features = ignored_features)
        model.fit(train_X, train_y)                                    # train on training set
        pred = model.predict_proba(test_X)                              # predict test set
        logloss = log_loss(test_y,pred)
        testScores['grid_'+str(i)].append(logloss)       # record test logloss/performance
        print('grid_'+str(i),num_ignored,logloss)
```

```python
# plotting logloss against number of least important features ignored
info_df = pd.concat([pd.DataFrame(testScores,index = range(0,24)),pd.Series(range(0,len(X.columns)-2,2), name = 'num_ignored')],axis=1)
for col_ in info_df.columns[:-1]:
    sb.lineplot(data = info_df.iloc[:-17], y = col_, x = 'num_ignored')
```

# Observation and Conclusion:
- The trend makes sense - as more features are ignored, there is lesser informaation, the model becomes more and more underfitted
- However, we can see that **dropping the first 2 (or 1) columns may lead to greater performance**
    - this shows that there are **columns are act as noise, dropping them will improve model generalizability**
